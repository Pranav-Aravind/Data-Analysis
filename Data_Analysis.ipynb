```python
import pandas as pd
import statistics
from matplotlib import pyplot as plt
import matplotlib.pyplot as pyplot

data = pd.read_csv(r'C:\Work\phone_battery_life.csv')

range_Amy = max(data['Amy']) - min(data['Amy'])
range_Brian = max(data['Brian']) - min(data['Brian'])

print('Battery life of Amys phone')
print('Mean: ', round(statistics.mean(data['Amy']),6))
print('Median: ', round(statistics.median(data['Amy']),6))
print('Range: ', round(range_Amy,6))
print('Standard Deviation: ', round(statistics.stdev(data['Amy']),6))

print('\nBattery life of Brians phone')
print('Mean: ', round(statistics.mean(data['Brian']),6))
print('Median: ', round(statistics.median(data['Brian']),6))
print('Range: ', round(range_Brian,6))
print('Standard Deviation: ', round(statistics.stdev(data['Brian']),6))

plt.hist((data['Amy']), alpha = 0.5, bins = 15)
plt.hist((data['Brian']), alpha = 0.5, bins = 15)
pyplot.title('Distribution of the time between charges over number of days for each phone')
pyplot.xlabel('Time between charges')
pyplot.ylabel('Number of days (out of 100 days)')
handles = ['Amys phone', 'Brians phone']
plt.legend(handles)
```

```python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt
import matplotlib.pyplot as pyplot

clicks = stats.poisson.rvs(25920, size=365)

cost = 2000 + clicks / 10

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(clicks,bins=15)
pyplot.title('Distribution of the number of clicks over number of days')
pyplot.xlabel('Number of clicks')
pyplot.ylabel('Number of days (out of 365 days)')

plt.subplot(1, 2, 2)
plt.hist(cost,bins=15,color='Red')
pyplot.title('Distribution of the cost over number of days')
pyplot.xlabel('Cost')
pyplot.ylabel('Number of days (out of 365 days) ')
```
